```python
# we are going to create a basic function but take into account that it can only take number so use try to add that if the program have a letter it sends an error message.
# The idea is to ask to put a number
try:
    a = int(input())
    b = int(input())
    c = int(input())
    # the three value that we need as an input (use the input as sting but we need number so transform them into integers as they are number)
    print(a + b + c)
    # print the result
except ValueError:
    print("Error: please enter valid numbers")
    # If something goes wrong, Python runs our error message instead
```

```python
# Second option with one single input
try:
    a, b, c = map(int, input("Enter 3 numbers separated by spaces: ").split())
    print(a + b + c)
except ValueError:
    print("Error: please enter valid numbers")
```
