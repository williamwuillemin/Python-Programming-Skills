```python
# Task 1
# please find a detail expalination. The first part is the part asked in the instruction and the rest is small addition to try if it worked

fh = open("hello.txt", "w")                                                         # open the document in mode write
lines_of_text = ["a line of text ", "another line of text ", "a third line \n"]     # type the text that we will later add just after (note that we need \n in order to have different line otherwise everything stays on the same line)
fh.writelines(lines_of_text)                                                        # add the line to the text
fh.close()                                                                          # close the document

fh = open("hello.txt", "a")                                                         # open the document in mode append
fh.write("Another new line of text.\nAnother line of text")                         # add two line using the append mode (not rewriting but adding)
fh.close()                                                                          # adding the document


fh = open("hello.txt", "r")                                                         # note that we can check the content but simply need to open it as read and not write
print(fh.read())                                                                    # print the content of the slide
fh.close()                                                                          # close the file after reading it
```
