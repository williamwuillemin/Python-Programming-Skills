# Python Skills Lesson 10 - File handling

```python
import os

print(os.getcwd())      # note that the file is saved into content (because we are in a googlecolab). This translate into is disapearance once the page is closed
```

```python
# Task 1
fh = open("hello.txt", "w")                                                         # open the document in mode write
lines_of_text = ["a line of text\n", "another line of text\n", "a third line\n"]    # type the text that we will later add just after (note that we need \n in order to have different line otherwise everything stays on the same line)
fh.writelines(lines_of_text)                                                        # add the line to the text
fh.close()                                                                          # close the document
```

```python
# checking the file by reading
fh = open("hello.txt", "r")
print(fh.read())
fh.close()
```
