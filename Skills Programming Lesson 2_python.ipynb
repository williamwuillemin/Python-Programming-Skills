```python
# Task 1
try:
    x = float(input("Please could you give me a number (it can contains digits): "))
    # define the inital input
    if x < 100:
        print("Your number is smaller than 100")
    else:
        print("Your number is higher than 100")
    # establish the two options
except ValueError:
    print("Error - you have to enter a float number")
    # simply handle the possible mistake
```

```python
# Other versions
try:
    number = float(input("Please could you give me a number (it can contains digits): "))
    if number < 100:
        print("Your number is smaller than 100")
    elif number == 100:
        print("Your number is equal to 100")
    else:
        print("Your number is higher than 100")
except ValueError:
    print("Error - you have to enter a float number")
```

```python
# Task 2
try:
  c = float(input("Please could you give me a temperature in Fahrenheit: "))    # Initial temperature
  f = (c - 32) * 5 / 9                                                          # Conversion formula
  print(f"The temperature in degree Celcius is {f}.")
except ValueError:
  print("Error, could you please enter a valid number")                         # Error message in case string instead of number
```

```python
# Other version
try:
    temperature = float(input("Enter the temperature: "))
    choice = input("To convert to Celsius put a C or to Fahrenheit with an F? ").upper()

    if choice == "C":
        result_a = (temperature - 32) * 5 / 9
        print(f"The termperature {temperature} in Fahrenheit equals {result_a} in Celicus")
    elif choice == "F":
        result_b = (temperature * 9 / 5) + 32
        print(f"The temperature {temperature}°C equal {result_b} in Fahrenheit")
    else:
        print("Invalid choice. Please enter C or F.")
except ValueError:
    print("Error: please enter a valid number.")
```

```python
# Task 3
try:
  mystring = input("Please could you write hello: ").lower()                    # no need to add str() as output already one
  myfloat = float(input("Please could you write 10.0: "))
  myint = int(input("Please could you write 20: "))                               # use int as if the user put 20.3, it directly sends the error message print valid input.
  if mystring == "hello":
    print(mystring)
  if myfloat == 10.0:
    print(myfloat)
  if myint == 20:
    print(myint)
except ValueError:
  print("Error, could you please enter a valid input.")
```

```python
# Task 4
# debug the below code so it works
# define the value
mystring = "hello"
myfloat = 10.0
myint = 20

# testing code
if mystring == "hello":
    print("String: %s" % mystring)                        # use %s as it is a string
if isinstance(myfloat, float) and myfloat == 10.0:        # if the value is a float and is equal to 10, then it print the result
    print("Float: %g" % myfloat)                          # use %g as it pick between scientific notation and fixed point float
if isinstance(myint, int) and myint == 20:
    print("Integer: %d" % myint)                          # simply display the integer (note that float would be rounded toward 0)
```

```python
# Task 5 corrections
try:
    # input
    vorname, nachname = input("Dear user, could you give me your first name and your surname (separated by a comma). Only alphabetical characters are accepted: ").split(",")
    # need to remove the spacing otherwise it return an error like with William, Wuillemin instead of William,Wuillemin
    vorname = vorname.strip()
    nachname = nachname.strip()

    # check for correct style of name
    if vorname.isalpha() and nachname.isalpha():
        print("Perfect, your name only contains letters.")
        # check if the name is correct
        vor_x = input(f"Is your first name: {vorname}? Please, reply by Yes or No: ").upper()
        nach_x = input(f"Is your last name: {nachname}? Please, reply by Yes or No: ").upper()
        # check for only the negative case (if it is correct no need to change) -> != is for not YES
        if vor_x != "YES":
            vorname = input("Could you write your correct first name: ")
        if nach_x != "YES":
            nachname = input("Could you write your correct last name: ")
        # Now that we have the correct name we can print output
        print(f"Hello, {vorname} {nachname}.")
    else:
        print("Invalid, it can only contain alphabetical characters.")
# if people only give a first name and not last name like with William instead of William,Wuillemin
except ValueError:
    print("Error, could you please enter a valid input (two names separated by a comma).")
```
