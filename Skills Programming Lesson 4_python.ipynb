```python
# Task 1
import array
# import the packages
try: # prepare in case someone put letter or technical sign $
  my_array = array.array("i", [int(x) for x in input("Enter a list of integer to put in the initial array. Note that it needs to be separated by a comma").split(",")])
  # create the array with the input of user
  print(f"The array is: {my_array}")
  # just check the array in order to see which element one want to delete
  value = int(input("Which element number do you want to delete?: "))
  # just add the conditional to make sure the element is in the array
  if value < 1 or value > len(my_array):
      print(f"Error: element number {value} does not exist in the array (valid range: 1-{len(my_array)})")
  else: # the right value can be delete as it is valid
      del my_array[value - 1]
      print(f"New array: {my_array} after the deletion of {value}")
except ValueError: # prepare in case someone put letter or technical sign $
  print("Please give a valid input.")
```
