```python
# Task 1

# we are asked to print the date of today, yesterday and tomorrow:

# In order to do this there are two options we either do everything without a package and simply raw code everything with function or we can use packages

# version without packages (this resemble a bit what we did in one of the previous exercise where we had to give the date).

# we first need to check if it is a leap year because then it is going to change what are the end of the months.
def is_leap_year(y):
    return y % 4 == 0 and (y % 100 != 0 or y % 400 == 0)

def days_in_month(m, y):
    return [31, 29 if is_leap_year(y) else 28, 31, 30, 31, 30,
            31, 31, 30, 31, 30, 31][m - 1]

def add_days(y, m, d, offset):
    d += offset
    while d > days_in_month(m, y):
        d -= days_in_month(m, y)
        m += 1
        if m > 12:
            m = 1
            y += 1
    while d < 1:
        m -= 1
        if m < 1:
            m = 12
            y -= 1
        d += days_in_month(m, y)
    return y, m, d

y, m, d = map(int, input("Enter today's date (YYYY-MM-DD): ").split("-"))

yesterday = add_days(y, m, d, -1)
tomorrow  = add_days(y, m, d,  1)

fmt = lambda y, m, d: f"{y}-{m:02d}-{d:02d}"

print(f"Yesterday : {fmt(*yesterday)}")
print(f"Today     : {fmt(y, m, d)}")
print(f"Tomorrow  : {fmt(*tomorrow)}")
```

```python
# Packages Import version

import datetime

today = datetime.date.today()
yesterday = today - datetime.timedelta(days=1)
tomorrow = today + datetime.timedelta(days=1)

print(f"Yesterday : {yesterday}")
print(f"Today     : {today}")
print(f"Tomorrow  : {tomorrow}")
```
