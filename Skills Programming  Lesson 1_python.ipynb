```python
# we are going to create a basic function but take into account that it can only take number so use try to add that if the program have a letter it sends an error message.
# The idea is to ask to put a number

try:                                                # safety net in case the program fails and return error (case with letter that cannot be transform in integer)
    a = int(input())                                # three input which are transform in integer
    b = int(input())
    c = int(input())
    # the three value that we need as an input (use the input as sting but we need number so transform them into integers as they are number)
    print(a + b + c)                                # print the result
except ValueError:                                  # second part of the safety net (handle the error)
    print("Error: please enter valid numbers")      # If something goes wrong, Python runs our error message instead
```

```python
try:                                                              # safety net to avoid program crashing
    k = int(input("Enter a day number between 1 and 365: "))      # input from the user
    if k < 1 or k > 365:                                          # conditional to verify the validity of the number of day
        print("Please enter a number between 1 and 365.")         # simply validity issue
# make sure we stay in the right year otherwise the months do not work anymore
    else:
        day_number = (k + 3) % 7                                  # check what the day is going to be
# add the + 3 as in the exercise the 1 of Jan was a thursday (so k = 1, + the 3 we have our 4)
# list of day
        days = [("Sunday", 0), ("Monday", 1), ("Tuesday", 2), ("Wednesday",3), ("Thursday",4), ("Friday",5), ("Saturday",6)]
# Days in each month in our case (non-leap year) so that we can find the data (note that it comes from the google calendar number of day)
# limit of each month
        months = [("Jan", 31), ("Feb", 28), ("Mar", 31), ("April", 30), ("May", 31), ("June", 30), ("July", 31), ("Aug", 31), ("Sep", 30), ("Oct", 31), ("Nov", 30), ("Dec", 31)]
        resid_k = k                                               # we need the residual in order to change from a month to the other
        for exact_m , number_d in months:                         # create a loop to go through all the list of month depending on the number of day remaining
          if resid_k <= number_d:                                 # case where we have the right number of day in the month
            break                                                 # we have our right months so we do not need to keep our loops
          else:                                                   # continue to go though the list
            resid_k = resid_k - number_d                          # change the number of day left so we can find the right number of days
        print(f"Day {k} of this year is {exact_m} {resid_k}, which is a {days[day_number][0]}.")      # print result
# note that we have to add the [0] because otherwise, we have the day and the number with it so we only need the first part of the tuple (note that we choose a tuple because the order is fixed and simpler than the dictionary where we would have to indicate the right key to go fetch the information)
except ValueError:                                                # handle error
    print("Error: please enter a valid number.")
# added in case someone put a word instead of number.
```
