```python
# Task 1
s = input("Please give me a sentence: ")                                                 # input sentence from the user
letter = input("Please give me a letter or number to count: ")                           # letter that need to be counted (same for the number)

word = (len(s.split()))                                                                  # split the word in a list and then only count the lenght of the list                                              # split the word and only count then the lenght of the list
print(f"There is {word} words in the sentence given.")                                         # return message

if len(letter) != 1:                                                                     # needed to add 1 otherwise one could put directly a word and not a letter so now it only count the letter (need a single letter to count)
    print("Please enter exactly one letter.")                                            # return message
else:                                                                                    # if the letter to count is valid then we count the number of letter
    letter_c = s.lower().count(letter.lower())                                           # transform all the letter in lowercase in order to be also able to count the first letter of a name for example. Then it count the number of the chosen letter
    print(f"There is {letter_c} {letter} in the sentence.")                              # return message

# note that we do not need any try / except because we only count number or strange sign in the sentence and they are all store as a string

# final output
print(f"The sentence countains {word} words and has {letter_c} letter {letter}.")         # simply the final output of the code so it looks nice
```

```python
# Task 2 (improvement of the initial output)
e = input("Enter the equation (such as 1 + 1 = 2): ")
# input equation with which we need to change the letter

# create a dictionary with all the number and change to operate
word = {"0": "zero", "1": "one", "2": "two", "3": "three", "4": "four", "5": "five", "6": "six", "7": "seven", "8": "eight", "9": "nine"}

e_sec = e.split()                                             # separate the different part of initial sentence

# create an empty entry that we are going to print with the new change
result = []
for x in e_sec:                                               # for loop to go for each element in the intial sentence
  if x.isdigit():
    translation = " & ".join(word[char] for char in x)        #  check if the number can be change
    result.append(translation)                                # add the new word to the empty set at the begining
  else:
    result.append(x)                                          # add the sign as sign like + , / , =
print(" ".join(result))                                       # print the final result

```
