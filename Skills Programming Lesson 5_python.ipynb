Skills Programming - Lesson 5 Strings

```python
# Task 1
s = input("Please give me a sentence: ")
letter = input("Please give me a letter or number to count: ")      # both input for the sentence and which letter do we want print(len(s.split()))

word = (len(s.split()))                                               # split the word and then only count the lenght of the list                                              # split the word and only count then the lenght of the list
print(word)

if len(letter) != 1:                                                # needed to add 1 otherwise one could put directly a word an not a letter so now it only count the letter
    print("Please enter exactly one letter.")
else:                                                               # transform all the letter in lowercase in order to be also able to count the first letter of a name for example
    letter_c = s.lower().count(letter.lower())
    print(letter_c)

# note that we do not need any try / except because we only count number or strange sign in the sentence and they are all store as a string

print(f"The sentence countains {word} words and has {letter_c} letter {letter}.")
```
