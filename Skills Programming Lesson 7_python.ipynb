```python
# Task 1
try:
  a = input("Please enter numbers (separated by a comma): ").split(",")
  numbers = []
  for item in a:
    item = item.strip()
    numbers.append(int(item))
  print(f"The sum of your input is {sum(numbers)}.")
except ValueError:
  print("Please enter numbers separated by a comma")
```

```python
# Task 1 (different interpretation)
a = input("Please enter numbers (separated by a comma): ").split(",")
total = 0
for item in a:
  try:
    item = item.strip()
    total += int(item)
  except ValueError:
    print(f"We are skipping the entry {item} as it is not a valid one")

print(f"The total sum of the numbers given is {total}")
```

```python
# Task 1 (other options)
a = input("Please enter numbers (separated by a comma): ").split(",")                         # takes the list of input and make sure they are split (using the comma)
total = 0                                                                                     # create the inital sum of value in which we are going to store the additional
for item in a:                                                                                # the loop that goes through each element given as input one by one
    item = item.strip()                                                                       # need to add this because the program consider the input as an invalid input (remove the spacing)
    while True:                                                                               # as long as the there is an input the loops keeps adding them but one there is none any more the loops stop
        try:                                                                                  # this is the part that makes sure that all the input are valid, before transforming the string into integer and summing them
            total += int(item)
            break                                                                             # valid input, move to next item
        except ValueError:
            item = input(f"The input {item} is not valid. Please enter a valid input: ")    # invalid input are isolated and the program asks to the user to replace the input in order to have a new compelete list. Note that it can also skip the item simply using: print(f"We are skipping the entry {item} as it is not a valid one")

print(f"The total sum of the numbers given is {total}")                                       # return the final sum
```

```python
# Task 2

def sumofthree():
# this is the definition of the function
  try:
# add the try: protection in order to protect the program if the user decide to separate by something else than a comma or if the user uses sign $ or characters instead of numbers
    a,b,c = input(f"Please enter three numbers (separated by a comma (Example: 1,2,3)): ").strip().split(",")
# this is the input function that ask the user to plug three numbers (note that another options would be to put the input directly inside the function such as sumofthree(1,2,3) but this version is more interactive)
# we remove the spacing and split the input by the comma
    numbers = []                                    # create an empty list in which we are going to put the transformed input
    for x in [a,b,c]:                               # create a for loop in order to be able to take both float and integers and obtain the sum of both (not only one type)
      if "." in x:                                  # we differentiate the two sort by the dot that only the float contain
        numbers.append(float(x))                    # when have a dot then we transform the input into a float
      else:
        numbers.append(int(x))                      # when it does not contain any dot then we know it is an integer
# note that if the input is not an integer or a float (the program cannot transform it and it does to the expect ValueError)
    sum = numbers[0] + numbers[1] + numbers[2]      # this is the sum of the three component of the transformed output
    print(f"The sum of {a}, {b} and {c} is {sum}.") # print the result
  except ValueError:                                # as mentioned is here in case of invalid input
    print("Error: Please enter three numbers separated by a comma (,)")




sumofthree()                                        # invoke the function
```

```python
# Task 2 (other simpler version)

def sumofthree(a,b,c):
# this is the definition of the function
  try:
# add the try: protection in order to protect the program if the user decide to separate by something else than a comma or if the user uses sign $ or characters instead of numbers
    numbers = []                                    # create an empty list in which we are going to put the transformed input
    for x in [a,b,c]:                               # create a for loop in order to be able to take both float and integers and obtain the sum of both (not only one type)
# note that we need to transform into a string otherwise the program is not able to identify the dot
      if "." in str(x):                             # we differentiate the two sort by the dot that only the float contain
        numbers.append(float(x))                    # when have a dot then we transform the input into a float
      else:
        numbers.append(int(x))                      # when it does not contain any dot then we know it is an integer
# note that if the input is not an integer or a float (the program cannot transform it and it does to the expect ValueError)
    sum = numbers[0] + numbers[1] + numbers[2]      # this is the sum of the three component of the transformed output
    print(f"The sum of {a}, {b} and {c} is {sum}.") # print the result
  except ValueError:                                # as mentioned is here in case of invalid input
    print("Error: Please enter three numbers separated by a comma (,)")

sumofthree(1,2,3 )
```
